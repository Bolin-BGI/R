## main program

```r
nFeature_RNA_low = 700
nFeature_RNA_high =6000
nCount_RNA_high=30000
percent.mt_up = 20

# txt.gz 格式读取
# rawdata <- fread(paste0(data_path, name),data.table = F)
# rownames(rawdata) <- rawdata$V1
# rawdata <- rawdata[,-1]
# head(rawdata)[1:5,1:2]

path <- paste0(data_path, sample,'/filtered_feature_bc_matrix/')
rawdata = Read10X(path, gene.column=2 ) # Attention： gene.column=2  联川生物

obj <- CreateSeuratObject(counts = rawdata, project = sample,min.cells = 5, min.features = 300)
obj <- RenameCells(obj, sample) # 每个细胞添加样本名前缀
obj[["percent.mt"]] <- PercentageFeatureSet(obj, pattern = "^MT-")


# VlnPlot-1
options(repr.plot.width = 10, repr.plot.height = 6)
p1 = VlnPlot(obj, features = c("nFeature_RNA"), ncol = 1, pt.size=0, y.max=3000) + geom_boxplot(width=.2,col="black",fill="white",outlier.size = 0.3) + NoLegend()
p2 = VlnPlot(obj, features = c("nCount_RNA"), ncol = 1, pt.size=0, y.max=9000) + geom_boxplot(width=.2,col="black",fill="white",outlier.size = 0.3) + NoLegend()
p3 = VlnPlot(obj, features = c("percent.mt"), ncol = 1, pt.size=0) + geom_boxplot(width=.2,col="black",fill="white",outlier.size = 0.3) + NoLegend()
combined_plot = (p1 | p2 | p3)
print(combined_plot)
ggsave(paste0(sample, "_unQC_vln.png"), combined_plot, width = 24, height = 8)

# print info
unqc_obj_info <- capture.output(obj)
unqc_mean_count <- mean(obj$nCount_RNA)
unqc_median_count <- median(obj$nCount_RNA)
cat("\n","过滤前:",unqc_obj_info)
cat("\n","Average counts per cell:", unqc_mean_count,"\n")
cat("Median counts per cell:", unqc_median_count,"\n")

##########################---subset---##########################
obj <- subset(obj,subset = nFeature_RNA > nFeature_RNA_low & 
                nFeature_RNA < nFeature_RNA_high & 
                nCount_RNA < nCount_RNA_high &
                percent.mt < percent.mt_up)

# print info 
qc_obj_info <- capture.output(obj)
qc_mean_count <- mean(obj$nCount_RNA)
qc_median_count <- median(obj$nCount_RNA)
cat("\n","过滤后:",qc_obj_info)
cat("\n","Average counts per cell:", qc_mean_count,"\n")
cat("Median counts per cell:", qc_median_count,"\n")               

# VlnPlot-2   
options(repr.plot.width = 10, repr.plot.height = 6)
p1 = VlnPlot(obj, features = c("nFeature_RNA"), ncol = 1, pt.size=0, y.max=3000) + geom_boxplot(width=.2,col="black",fill="white",outlier.size = 0.3) + NoLegend()
p2 = VlnPlot(obj, features = c("nCount_RNA"), ncol = 1, pt.size=0, y.max=9000) + geom_boxplot(width=.2,col="black",fill="white",outlier.size = 0.3) + NoLegend()
p3 = VlnPlot(obj, features = c("percent.mt"), ncol = 1, pt.size=0) + geom_boxplot(width=.2,col="black",fill="white",outlier.size = 0.3) + NoLegend()
combined_plot = (p1 | p2 | p3)
print(combined_plot)
ggsave(paste0(sample, "_QC_vln.png"), combined_plot, width = 24, height = 8)
                
                
##########################---Find_doublet---##########################
Find_doublet <- function(data,ratio = 0.05){
        sweep.res.list <- paramSweep_v3(data, PCs = 1:10, sct = FALSE)
        sweep.stats <- summarizeSweep(sweep.res.list, GT = FALSE)
        bcmvn <- find.pK(sweep.stats)
        nExp_poi <- round(as.numeric(ratio)*ncol(data)) # ratio：预期双胞的比例，例如 0.05 表示 5% 的细胞可能是双胞
        p <- as.numeric(as.vector(bcmvn[bcmvn$MeanBC==max(bcmvn$MeanBC),]$pK))
        data <- doubletFinder_v3(data, PCs = 1:10, pN = 0.25, pK = p, nExp = nExp_poi, reuse.pANN = FALSE, sct = FALSE)
        colnames(data@meta.data)[ncol(data@meta.data)] = "doublet_info"
        data
}

obj <- obj %>% NormalizeData %>%  FindVariableFeatures %>% ScaleData %>% RunPCA %>% RunUMAP(1:30) %>% Find_doublet

obj$doublet_score = obj@meta.data[,5]   # 有 bug!
p1 = DimPlot(obj, group.by='doublet_info') + NoLegend()
ggsave(paste0(sample, "_QC_doublet.png"), p1, width=7)         
p2 = FeaturePlot(obj, "doublet_score")
ggsave(paste0(sample, "_QC_doublet.png"), p1 + p2, width=14)

obj$name = name
obj$sample = sample
obj$type = type
obj$GEO = GEO
     
obj <- obj %>%
  subset(doublet_info=="Singlet") %>%
  NormalizeData %>%  
  FindVariableFeatures %>%
  ScaleData %>%
  RunPCA %>% 
  RunUMAP(1:30) %>%
  FindNeighbors %>%
  FindClusters

p1 = DimPlot(obj, label=T) + NoLegend()
p2 = FeaturePlot(obj, "percent.mt")
p3 = FeaturePlot(obj, "nFeature_RNA")
p4 = FeaturePlot(obj, "nCount_RNA")
ggsave(paste0(sample, "QC_umap.png"), (p1 | p2)/(p3 | p4))

              
########################---保存rds及结果---#######################                    
obj <- saveRDS(obj, paste0(sample, ".rds"))       
print("rds保存成功！")

########################---cat to file---#######################                
file_name <- file(paste0(sample,"_info.txt"), open = "w")
    cat("\n过滤前:\n", unqc_obj_info, file = file_name)
    cat("\n","Average counts per cell:", unqc_mean_count,"\n", file = file_name, append = TRUE)
    cat("Median counts per cell:", unqc_median_count,"\n", file = file_name, append = TRUE)

    cat("\n过滤后:\n", qc_obj_info, file = file_name)
    cat("\nAverage counts per cell:", qc_mean_count, "\n", file = file_name, append = TRUE)
    cat("Median counts per cell:", qc_median_count, "\n", file = file_name, append = TRUE)
                
    # cat('\npig cell 范围：',range(obj$percent.p, na.rm = TRUE, finite = TRUE), file = file_name, append = TRUE)
    # cat('\nmonkey cell 范围：',range(obj$percent.m, na.rm = TRUE, finite = TRUE),"\n", file = file_name, append = TRUE)
close(file_name)
```
